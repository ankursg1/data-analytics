Automation for this

total unique purchase by the age group

```python
for i in df['Age'].unique():
    print('Age' ,'|' , 'Purchase Number')
    print(i,' : ',len(df[df['Age']== i]['Product_ID'].unique()))
```

total purchase by the age group including same item

```python
for i in df['Age'].unique():
    print('Age' ,'|' , 'Purchase Number')
    print(i,' : ',len(df[df['Age']== i]))
```

```python
lst=[]
for i in df['Age'].unique():
    lst.append([i,df[df['Age']==i]['Product_ID'].nunique()])

data = pd.DataFrame(lst,columns =["Age",'Product'])
```

```python
data
```

```python
data.plot(kind='bar')
```

```python
data.plot.bar(x='Age', title='Amount Distribution by Age')
```
