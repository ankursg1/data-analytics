# To plot it on pi-chart

figsize() --> to increase the size of the figure.

autopct() ---> for adding percentage in pie-chart (1f)==til 1 decimal place

```python
data=pd.DataFrame({'Ratio' : [len(df[df['Gender']=='M']) , len(df[df['Gender']=='F'])]},index=('Male','Female'))
data.plot.pie(y='Ratio',figsize=(7,7), autopct='%.1f')
```

## OR

```python
df.groupby('Gender').size().plot(kind='pie' , autopct='%.1f' , title='Gender Ratio' , figsize=(7,7))
```

```python
df.groupby('Gender').size().plot(kind ='bar' , figsize=(7,7))
```

```python
df.groupby('Gender').sum()['Purchase'].plot(kind='pie',  autopct='%.1f')
```

## how much money male or female are spending per product

```python
df.groupby('Gender').mean()['Purchase'].plot(kind='pie',  autopct='%.1f')
```
